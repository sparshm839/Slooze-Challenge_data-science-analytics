## ABC Analysis


```python

# Inventory Value per item
begin_final['InventoryValue'] = begin_final['onHand'] * begin_final['Price']

abc_df = begin_final.groupby(['Brand', 'Description'], as_index=False)['InventoryValue'].sum()

abc_df = abc_df.sort_values(by='InventoryValue', ascending=False).reset_index(drop=True)


abc_df['CumulativeValue'] = abc_df['InventoryValue'].cumsum()
total_value = abc_df['InventoryValue'].sum()
abc_df['CumulativePerc'] = 100 * abc_df['CumulativeValue'] / total_value

# Diving into ABC
def classify_abc(p):
    if p <= 70:
        return 'A'
    elif p <= 90:
        return 'B'
    else:
        return 'C'

abc_df['ABC_Class'] = abc_df['CumulativePerc'].apply(classify_abc)

# Exporting ABC Class image
abc_df.head(10)  
```

```python
a_items = abc_df[abc_df['ABC_Class'] == 'A'][['Brand', 'Description']]
print(" A-Class Inventory (High Value):")
print(a_items)
b_items = abc_df[abc_df['ABC_Class'] == 'B'][['Brand', 'Description']]
print(" B-Class Inventory (Moderate Value):")
print(b_items)
c_items = abc_df[abc_df['ABC_Class'] == 'C'][['Brand', 'Description']]
print(" C-Class Inventory (Low Value):")
print(c_items)
```

## Economic Order Quantity (EOQ) Analysis

```python
# EOQ = root(2DS/H)
# D = Annual demand (units sold per year)
# S = Ordering cost per order
# H = Holding (carrying) cost per unit per year

import pandas as pd
import numpy as np


sales_summary = sales_final.groupby('InventoryId')['SalesQuantity'].sum().reset_index()
sales_summary.columns = ['InventoryId', 'AnnualDemand']

eoq_df = pd.merge(end_final, sales_summary, on='InventoryId', how='inner')

ordering_cost = 500  # fixed cost per order
eoq_df['HoldingCost'] = 0.1 * eoq_df['Price']  # 10% of unit price

# EOQ
eoq_df['EOQ'] = np.sqrt((2 * eoq_df['AnnualDemand'] * ordering_cost) / eoq_df['HoldingCost'])


eoq_df['EOQ'] = eoq_df['EOQ'].round()

eoq_output = eoq_df[['InventoryId', 'Brand', 'Description', 'AnnualDemand', 'Price', 'HoldingCost', 'EOQ']]

eoq_output.head(10)
```

##   Reorder Point Analysis


```python
# Reorder Point (ROP) = Average Daily Demand  Lead Time (in days)
#                      => Safety Stock
```

```python
import pandas as pd
import numpy as np


sales_final['SalesDate'] = pd.to_datetime(sales_final['SalesDate'])
daily_sales = sales_final.groupby(['InventoryId', 'SalesDate'])['SalesQuantity'].sum().reset_index()


demand_stats = daily_sales.groupby('InventoryId')['SalesQuantity'].agg(['mean', 'std']).reset_index()
demand_stats.rename(columns={'mean': 'avg_daily_demand', 'std': 'std_daily_demand'}, inplace=True)


LEAD_TIME_DAYS = 7
Z_FACTOR = 1.65


demand_stats['safety_stock'] = Z_FACTOR * demand_stats['std_daily_demand'] * np.sqrt(LEAD_TIME_DAYS)
demand_stats['reorder_point'] = (demand_stats['avg_daily_demand'] * LEAD_TIME_DAYS) + demand_stats['safety_stock']


reorder_df = pd.merge(demand_stats, end_final[['InventoryId', 'Description', 'Price']], on='InventoryId', how='left')
reorder_df = reorder_df.sort_values(by='reorder_point', ascending=False)


reorder_df[['InventoryId', 'Description', 'avg_daily_demand', 'safety_stock', 'reorder_point']].head(10)
```

## Statistical Visualization for ROP

```python
# Bar Chart-Top 10ROP products
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_rop,
    x='reorder_point',
    y='Description',
    hue='Description',
    dodge=False,
    palette='crest',
    legend=False  # hide legend since each bar is labeled
)
plt.title('Top 10 Products by Reorder Point')
plt.xlabel('Reorder Point')
plt.ylabel('Product Description')
plt.tight_layout()
plt.savefig("../visuals/reoder_point_plot.png")
plt.show()

```

## Lead Time Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

invoice_purchase['InvoiceDate'] = pd.to_datetime(invoice_purchase['InvoiceDate'])
invoice_purchase['PODate'] = pd.to_datetime(invoice_purchase['PODate'])

invoice_purchase['LeadTime_Days'] = (invoice_purchase['InvoiceDate'] - invoice_purchase['PODate']).dt.days

vendor_leadtime = invoice_purchase.groupby(['VendorNumber', 'VendorName'])['LeadTime_Days'].mean().reset_index()
vendor_leadtime = vendor_leadtime.sort_values(by='LeadTime_Days', ascending=False)

# top vendors with highest average lead time
vendor_leadtime.head(10)
```

## Statistical Visualization for Lead Time Analysis

```python

import matplotlib.pyplot as plt
import seaborn as sns

top_vendors = vendor_leadtime.sort_values('LeadTime_Days', ascending=False).head(10)


top_vendors['DummyHue'] = 'All'

plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_vendors,
    x='LeadTime_Days',
    y='VendorName',
    hue='DummyHue',               # Assign hue
    palette='flare',
    dodge=False                   # Don't split bars
)
plt.title('Top 10 Vendors by Average Lead Time')
plt.xlabel('Average Lead Time (Days)')
plt.ylabel('Vendor Name')
plt.legend([], [], frameon=False)  # Remove legend
plt.tight_layout()
plt.savefig("../visuals/lead_time_vendors.png")  
plt.show()
```
